# Jul. 15: dimensionality reduction

The overall performance is not so good: the "blocks" in the dimreduct graph has no strong relationship with the conformation. After inspection on the descriptors, it turned out that the dihedral descriptors have two well-separated peaks that are symmetrical to the y-axis. This is because the dihedral ranges from -180 to 180 degrees, so both positive and negative values occur. The varience between the peaks are so huge that the dimreduct algorithm fails to recognize other variations in the descriptors. 

We also noticed that the peak is quite narrow, indicating that the variance on either side are quite small. We plan to solve the problem of poor performance by taking absolute value when calculating the dihedrals.
